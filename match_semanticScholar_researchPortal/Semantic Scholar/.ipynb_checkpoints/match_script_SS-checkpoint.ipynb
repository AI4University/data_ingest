# 2. Obtain the relationships of the IDs of Research Portal and Semantic Scholar

## 2.1. Check how many Research Portal publications can we find in Semantic Scholar

```python
df_semantic_publications = df_semantic_publications.withColumn("tituloNormalizado_SS", normalize_title_udf(df_semantic_publications["title"]))
df_semantic_publications.show(n=1, truncate=120, vertical=True)
```

```python
df_publications = df_publications.select('actID', 'Title') 
df_publications = df_publications.withColumn("tituloNormalizado_RP", normalize_title_udf(df_publications["Title"]))
df_publications.show(n=1, truncate=120, vertical=True)
```

```python
df_matched_title = df_publications.join(df_semantic_publications, 
                                   df_publications["tituloNormalizado_RP"] == df_semantic_publications["tituloNormalizado_SS"], 
                                   'inner').cache()

df_matched_title = df_matched_title.dropDuplicates(['actID']).dropDuplicates(['id'])
df_publis_ids = df_matched_title.select(col("actID").alias("ID_RP"), col("id").alias("ID_SS"))

print('Total of publications in the Research Portal:', df_publications.count())
print('Number of publications matched by title after removing duplicates:', df_matched_title.count())
```

```python
df_matched_title = df_matched_title.select('actID', 'id')
```

## 2.2. Find the relationship between the researchers IDs of the research portal and Semantic Scholar

```python
print('Signatures before cleaning in Semantic Scholar:', df_semantic_publication_author.count())

df_semantic_publication_author_resumido = df_semantic_publication_author.join(df_publis_ids, 
                                   df_semantic_publication_author["paper_id"] == df_publis_ids["ID_SS"], 
                                   'inner').cache()

print('Signatures after cleaning in Semantic Scholar:', df_semantic_publication_author_resumido.count())
```

```python
print('Authors before cleaning in Semantic Scholar:', df_semantic_authors.count())

df_semantic_authors_resumido = df_semantic_authors.join(df_semantic_publication_author_resumido, 
                                   df_semantic_publication_author_resumido["author_id"] == df_semantic_authors["id"], 
                                   'inner').cache()

print('Authors after cleaning in Semantic Scholar:', df_semantic_authors_resumido.count())
df_semantic_authors_resumido = df_semantic_authors_resumido.select('id', 'name', 'papercount', 'aliases')
df_semantic_authors_resumido = df_semantic_authors_resumido.drop_duplicates()
print('Authors after removinig duplicates in Semantic Scholar:', df_semantic_authors_resumido.count())
```

```python
new_lines = match_name_SS_RP(match_publis_SS_RP=df_publis_ids, researchers_publications_RP=df_researchers_publications, researchers_publications_SS=df_semantic_publication_author_resumido,
                             researchers_RP=df_researchers, researchers_SS=df_semantic_authors_resumido)

# Crear DataFrame a partir de la lista de objetos Row y lo guardo para ahorrar tiempo 
df = spark.createDataFrame(new_lines)
df_pd = df.toPandas()
df_pd.to_parquet('/export/usuarios_ml4ds/mbalairon/github/data_ingest/match_semanticScholar_researchPortal/output/df.parquet', index=False)
```

## 2.3. Identify more authors using the Alias

```python
# Create a new dataframe containig all the possible names for each of the researchers al ready matched 
new_rows = []
total_iter = df_ids_researchers.shape[0]
for i in tqdm(range(total_iter), desc="Procesando"):
    invID = df_ids_researchers['id_RP'][i]
    aliases = df_ids_researchers['aliases'][i]
    for alias in aliases:
        new_rows.append((invID, alias))  # Asegúrate de empaquetar invID y alias en una tupla

df_aliases_pd = pd.DataFrame(new_rows, columns=['id_RP', 'name_SS'])

# Convert the datasets to pySpark tables
schema = StructType([
    StructField('id_RP', StringType(), True),
    StructField('name_SS', StringType(), True),
    ])

rows = [Row(**record) for record in df_aliases_pd.to_dict(orient='records')]
df_aliases = spark.createDataFrame(df_aliases_pd, schema=schema)
```

```python
# Match the new names of the authors and look how many publications have we retrieved
df_match_aliases = df_semantic_authors.join(df_aliases, df_semantic_authors["name"] == df_aliases["name_SS"], 'inner').cache()
df_match_aliases = df_match_aliases.select(col("id").alias("id_SS"), "id_RP", 'name_SS', col('papercount').alias('no_publis_SS'))
print('New registered authors:', df_match_aliases.count())
```

```python
df_match_aliases_pd = df_match_aliases.toPandas()
df_match_aliases_pd.head()
```

```python
# junto ambos dataframes y los guardo
df_ids_researchers = df_ids_researchers[['name_RP', 'name_SS', 'id_RP', 'id_SS', 'no_publis_RP', 'no_publis_SS']]

df_more_ids_researchers = pd.concat([df_ids_researchers, df_match_aliases_pd], ignore_index=True)
df_more_ids_researchers['no_publis_RP'] = df_more_ids_researchers['no_publis_RP'].fillna(0)
df_more_ids_researchers['no_publis_SS'] = df_more_ids_researchers['no_publis_SS'].fillna(0)
df_more_ids_researchers['name_RP'] = df_more_ids_researchers.groupby('id_RP')['name_RP'].transform(lambda x: x.ffill().bfill())
df_more_ids_researchers = df_more_ids_researchers.sort_values(by='id_RP')
df_more_ids_researchers.to_parquet('/export/usuarios_ml4ds/mbalairon/github/data_ingest/match_semanticScholar_researchPortal/output/df2.parquet')
df_more_ids_researchers.head()
```

```python
# sumo todas las publicaciones para un mismo autor
df_plot_publis = df_more_ids_researchers[['name_RP', 'id_RP', 'no_publis_RP', 'no_publis_SS']].groupby(['id_RP', 'name_RP']).agg({
    'no_publis_RP': 'sum',
    'no_publis_SS': 'sum'
}).reset_index()
```
